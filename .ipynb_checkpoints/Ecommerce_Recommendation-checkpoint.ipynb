**Importing necessary libraries**

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as splt
import seaborn as sns

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import coo_matrix
```

**Loading the dataset**

```python
# Load your dataset 
train_data = pd.read_csv('C:/Users/USER/SLIIT/Y3S1 resources/z IRWA Project/amazon.csv')
train_data.head()
```

**Handling null, duplicates, data types**

```python
# Checking for null values
train_data.isnull().sum()
```

```python
# Filling the null values of 'rating_count' column with 0
train_data['rating_count'] = train_data['rating_count'].fillna(0) 
```

```python
train_data.isnull().sum()
```

```python
# Checking for duplicates
train_data.duplicated().sum()
```

```python
# Checking for the data types
train_data.dtypes
```

```python
train_data.head()
```

```python
# Object data type might not be efficient or suitable for certain operations
# especially when performing numerical computations, sorting, or building machine learning models

# Convert 'discounted_price' (if it contains currency symbols, remove them first)
train_data['discounted_price'] = train_data['discounted_price'].replace('[₹,]', '', regex=True)
train_data['discounted_price'] = pd.to_numeric(train_data['discounted_price'], errors='coerce')

# Convert 'actual_price' by removing currency symbols and commas, then convert to numeric
train_data['actual_price'] = train_data['actual_price'].replace('[₹,]', '', regex=True)
train_data['actual_price'] = pd.to_numeric(train_data['actual_price'], errors='coerce')

# Convert 'discount_percentage' by removing the '%' sign and converting to numeric
train_data['discount_percentage'] = train_data['discount_percentage'].str.replace('%', '', regex=False)
train_data['discount_percentage'] = pd.to_numeric(train_data['discount_percentage'], errors='coerce')

# Convert 'rating' to float
train_data['rating'] = pd.to_numeric(train_data['rating'], errors='coerce') 

# Convert 'rating_count' to integer
train_data['rating_count'] = pd.to_numeric(train_data['rating_count'], errors='coerce').fillna(0).astype(int)  
```

```python
train_data.dtypes
```

**Renaming columns**

```python
column_name_mapping = {
    'product_id': 'ProductID',
    'product_name': 'ProductName',
    'category': 'Category',
    'discounted_price': 'DiscountedPrice',
    'actual_price': 'ActualPrice',
    'discount_percentage': 'DiscountedPercentage',
    'rating': 'Rating',
    'rating_count': 'RatingCount',
    'about_product': 'AboutProduct',
    'user_id': 'UserID',
    'user_name': 'UserName',
    'review_id': 'ReviewID',
    'review_title': 'ReviewTitle',
    'review_content': 'Review',
    'img_link': 'ImageLink',
    'product_link': 'ProductLink'
}

train_data.rename(columns=column_name_mapping, inplace=True)
train_data.head()
```
